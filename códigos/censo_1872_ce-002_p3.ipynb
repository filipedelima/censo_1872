# População menor que 1 ano

```python
meses = df[['Unnamed: 0','Unnamed: 1', 'Unnamed: 2', 'Unnamed: 3', 'Unnamed: 4', 'Unnamed: 5', 'Unnamed: 6', 'Unnamed: 7', 'Unnamed: 8', 'Unnamed: 9', 'Unnamed: 10', 'Unnamed: 11', 'Unnamed: 12', 'Unnamed: 13']]
meses.head()
```

```python

mulheres_livres = df_no_transpose.loc[(df_no_transpose[('', 'Condições')] == 'Livres') & (df_no_transpose[('', 'Sexos')] == 'Mulheres')]
mulheres_escravas = df_no_transpose.loc[(df_no_transpose[('', 'Condições')] == 'Escravas') & (df_no_transpose[('', 'Sexos')] == 'Mulheres')]

soma_mulheres = mulheres_livres[colunas_bebes].sum() + mulheres_escravas[colunas_bebes].sum()

soma_mulheres
```

```python
total = soma_homens + soma_mulheres
total
```
