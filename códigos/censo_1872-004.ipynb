# Sexos

```python
sexos = df[['Unnamed: 0','Unnamed: 1', 'Unnamed: 2', 'Unnamed: 3']]
sexos.head()
```

```python
new_header = sexos.iloc[1]
sexos.columns = new_header
sexos = sexos[2:]
sexos.head()
```

```python
columns_multiindex = pd.MultiIndex.from_tuples([
    #('Gêneros', 'Províncias e Município Neutro'),
    ('', 'Províncias e Município Neutro'),
    ('Gêneros', 'Homens'),
    ('Gêneros', 'Mulheres'),
    ('Gêneros', 'Total')
])

# Atribuindo o MultiIndex às colunas do DataFrame
sexos.columns = columns_multiindex

#print(sexos)
sexos
```

```python
colunas_sexos = [('Gêneros', 'Homens'), ('Gêneros', 'Mulheres'), ('Gêneros', 'Total')]
sexos[colunas_sexos] = sexos[colunas_sexos].apply(pd.to_numeric)

total_homens = sexos.iloc[21,1]
total_mulheres = sexos.iloc[21,2]
total = sexos.iloc[21,3]

soma_homens = sexos[('Gêneros', 'Homens')].sum() - total_homens
soma_mulheres = sexos[('Gêneros', 'Mulheres')].sum() - total_mulheres
soma_total = sexos[('Gêneros', 'Total')].sum() - total


if soma_homens == total_homens:
    soma_homens
else:
    print("Discrepância entre valor digitado na tabela e soma dos valores")
    print("Digitado na tabela: " + str(total_homens))
    print("Valor da soma: " + str(soma_homens))
    print("Diferença: " + str(soma_homens - total_homens))

if soma_mulheres == total_mulheres:
    total_mulheres
else:
    print("Discrepância entre valor digitado na tabela e soma dos valores")
    print("Digitado na tabela: " + str(total_mulheres))
    print("Valor da soma: " + str(soma_mulheres))
    print("Diferença: " + str(soma_mulheres - total_mulheres))

if soma_total == total:
    total
else:
    print("Discrepância entre valor digitado na tabela e soma dos valores")
    print("Digitado na tabela: " + str(total))
    print("Valor da soma: " + str(soma_total))
    print("Diferença: " + str(soma_total - total))
```
